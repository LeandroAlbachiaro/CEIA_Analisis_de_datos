16. What is the average revenue amount per order?

```python
# Sumar la columna 'item_price' por order_id
ganancia_por_orden = chipo.groupby('order_id')['item_price'].sum()

# Calcular la ganancia promedio por orden
ganancia_promedio_por_orden = ganancia_por_orden.mean()
print(ganancia_promedio_por_orden)

#Otra forma
chipo['item_price'].sum()/chipo['order_id'].nunique()
```

```python
# agrupar por order_id y obtener la descripción de item_price
chipo.groupby('order_id')['item_price'].describe()
```
